```python
import pandas as pd
import re
```

```python
input_file = r'D:\School\ADMU\4Y\SEM 1\PSYC 80.18i\Project\Influencer Apologies - Sheet1.csv'
output_file = r'D:\School\ADMU\4Y\SEM 1\PSYC 80.18i\Project\Cleaned_Influencer_Apologies.csv'
text_file = r'D:\School\ADMU\4Y\SEM 1\PSYC 80.18i\Project\text.txt'
data = pd.read_csv(input_file)


# Clean the text by removing timestamps, [Music], [Applause], and extra whitespace
def clean_text(text):
    text = re.sub(r'\d{1,2}:\d{2}', '', str(text))  # Remove timestamps
    text = re.sub(r'\[Music\]|\[Applause\]', '', text)  # Remove [Music] and [Applause]
    text = re.sub(r',', '', text)
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra whitespace
    return text

data['Text'] = data['Text'].apply(clean_text)

# Read the text file
with open(text_file, 'r') as file:
    new_data = file.read()
display(new_data)
# Clean the new data
cleaned_new_data = clean_text(new_data)

# Replace NaN values in the 'Cleaned_Text' column with the cleaned new data
data.at[17,'Text'] = cleaned_new_data
# Save the updated data back to the CSV file
data.to_csv(output_file, index=False)
display(data)
```
